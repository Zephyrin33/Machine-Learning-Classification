```python
import pandas as pd
```

```python
dataset = pd.read_csv("CKD.csv")
```

```python
dataset
```

```python
dataset = pd.get_dummies(dataset,dtype=int,drop_first=True)
```

```python
dataset
```

```python
dataset.columns
```

```python
indep =  dataset[['age', 'bp', 'al', 'su', 'bgr', 'bu', 'sc', 'sod', 'pot', 'hrmo', 'pcv',
       'wc', 'rc', 'sg_b', 'sg_c', 'sg_d', 'sg_e', 'rbc_normal', 'pc_normal',
       'pcc_present', 'ba_present', 'htn_yes', 'dm_yes', 'cad_yes',
       'appet_yes', 'pe_yes', 'ane_yes']]
```

```python
indep
```

```python
dep =  dataset[['classification_yes']]
```

```python
dep
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test = train_test_split(indep,dep,test_size=0.30,random_state=0)
```

```python
y_train
```

```python
dataset["classification_yes"].value_counts()
```

```python
from sklearn.preprocessing import StandardScaler
sc = StandardScaler()
x_train = sc.fit_transform(x_train)
x_test = sc.transform(x_test)
```

```python
from sklearn.naive_bayes import ComplementNB
from sklearn.model_selection import GridSearchCV
param_grid = {'alpha': [0.1, 0.5, 1.0, 2.0],'fit_prior': [True, False],'norm': [True, False]} 
grid = GridSearchCV( ComplementNB(),param_grid,refit=True,verbose=3,cv=5,n_jobs=-1,scoring='f1_weighted')
grid.fit(x_train,y_train)
```

```python
#print(grid.best_params_)
re = grid.cv_results_
grid_predictions = grid.predict(x_test)
```

```python
from sklearn.metrics import confusion_matrix
cm = confusion_matrix(y_test,grid_predictions)
from sklearn.metrics import classification_report
clf_report = classification_report(y_test,grid_predictions)
print("The result of CM:\n",cm)
print("The result of clf_report:\n",clf_report)
```

```python
from sklearn.metrics import f1_score
f1_macro = f1_score(y_test,grid_predictions,average='weighted')
print('The f1_macro value for the best parameter{}:'.format(grid.best_params_),f1_macro)
```

```python
from sklearn.metrics import roc_auc_score
roc_auc_score(y_test,grid.predict_proba(x_test)[:,1])
```
